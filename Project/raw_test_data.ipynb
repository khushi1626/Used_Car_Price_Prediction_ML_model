```python
brand = []
model = []
for i in df['Name']:
    l=[]
    try:
        l = i.split(" ")
        if(l[0]=="Land"):
            brand.append(l[0] + " " + l[1])
            model.append(l[2])
        else:
            brand.append(l[0])
            model.append(l[1])
    except:
        continue
    
```
